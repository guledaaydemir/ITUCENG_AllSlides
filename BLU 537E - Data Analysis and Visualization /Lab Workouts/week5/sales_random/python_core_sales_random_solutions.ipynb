## Find out who paid the highest amount of money 

For this taks use  following exchange rates
<br> 1$=6tl, 1=8tl, 1=7tl

create list where each element is a list of payment in tl,customer id and customer name
<br> name this list as payment_list


first, write a function that takes input as string form like "100$" and returns floating point in tl
<br> name this function exchange

```python
def exchange(x):
    #find the currency
    currency=x[-1]
    #calculate the payment in tl
    if currency=="$":
        payment=6*float(x[:-1])
    elif currency=="":
        payment=8*float(x[:-1])
    elif currency=="":
        payment=7*float(x[:-1])
    return payment
        
```

test this function
<br> check what this  exchange("10$") returns

```python
print(exchange("10$"))
```

now we can form our payment_list

```python
#create an empty list
payment_list=[]
#go through lines list and take customer id ,name and payment
#put the payment string into the exchange function and return payment in tl
#note that there is a "\n" character in the payment you have to get rid of it
for line in lines:
    customer=line.strip().split(",")
    customer_id=customer[0]
    customer_name=customer[1]
    payment=exchange(customer[-1])
    payment_list.append([payment,customer_id,customer_name])
    
```

print payment_list

```python
payment_list
```

sort the list and find the customer who paid the highest amount

```python
payment_list.sort(reverse=True)
```

```python
print(payment_list[0][1],payment_list[0][2])
```

we got the area codes, now we can fill the missing city data information
<br> before starting, let's write a function that takes phone string and returns areacode
<br> simply we just put the scripts above into a function
<br> call this function get_area_code

```python
def get_area_code(x):
    digits=re.findall("\d",x)
    if digits[0]=="0":
        areacode=int("".join(digits[1:4]))
    else:
        areacode=int("".join(digits[0:3]))
    return areacode
```
