# Basic Plotting with Pyplot

First, import necessary libraries

```python
import matplotlib
import matplotlib.pyplot as plt
```

The plot command expects two arguments (Python list, NumPy array or pandas DataFrame) for the x and y axis respectively.

```python
plt.plot([1,2,4])
```

The latest versions of Jupyter Notebook seem to display the figure without calling plt.show() command explicitly. <br> However, in Python shell or embedded applications, plt.show() or plt.draw() is required to display the figure on the screen.

check the backend

```python
plt.rcParams['backend']
```

inline backend renders figures in the notebook as static pngs.

```python
y=2*x;y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,z)
```
