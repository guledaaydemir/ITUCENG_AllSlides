# Grouping

• The metacharacters ( and ) denote a group.
<br> Sometimes we would like to extract a part of the pattern not the whole pattern.
<br> in these cases we can use grouping

The parenthesis ( ) group mechanism can be combined with findall().
<br> • If the pattern includes 2 or more parenthesis groups, then instead of returning a list of strings, findall() returns a list of *tuples*.

```python
import re
```

```python
text="john-dentist-1984,marry-teacher-1986,jessica-nurse-2001,tony-doctor-1956"
```

let say we have names, occupations and birthdates as given above

we want to extract names and birthdates and print the ages of the people

first, extract names and birthdates

```python
result=re.findall("\w+-\w+-\d+",text)
```

```python
result
```

```python
result[0]
```

now only extract name and birthdate not the occupation

```python
result=re.findall("(\w+)-\w+-(\d+)",text)
```

```python
result
```

```python
result[0]
```

so we have a list where each element is a tuple containing name and birthdate

we can go through all the people in this list and print the ages

```python
for (name,year) in result: 
    print(name,year)
```

```python
for (name,year) in result: 
    print(name,2019-year)
```

```python
for (name,year) in result: 
    print(name,2019-int(year))
```

```python
for match in re.finditer("(\w+)-\w+-(\d+)",text):
    print(match)
```

```python
match.groups()
```

```python
match.group(0)
```

```python
match.group(1)
```

### Preceding and following matches

:

Here is an example that matched the word the only if it is followed by cat:

```python
re.sub(r'the(?= cat)', '*', 'the dog and the cat')
```

Here is an example that matches the word the only if it is preceded by a space:

```python
re.sub(r'(?<= )the', '*', 'Athens is the capital.')
```

The following example will match the word the only if it neither preceded by and nor followed by letters, so you can use it to replace occurrences of the word the, but not occurrences of the within other words.

```python
re.sub(r'(?<!\w)[Tt]he(?!\w)', '*', 'The cat is on the lathe there.')
```

# Formatting integers

To format integers, the formatting code is {:d}.

Putting a number in front of the d allows us to right-justify integer

<, >, =, or ^, for left alignment, right alignment, padding after a sign character, or centered alignment

```python
print('{:3d}'.format(2)) 
print('{:3d}'.format(25)) 
print('{:3d}'.format(138))
print('{:<3d}'.format(2))
print('{:03d}'.format(2)) 
print('{:03d}'.format(25)) 
print('{:8d}'.format(25))
print('{:<8d}'.format(25))
print('{:^8d}'.format(25))
```

# Formatting Floats

To format a floating point number, the formatting code is {:f}.

To only display the number to two decimal places, use {:.2f}.

```python
print('{:6.2f}'.format(3.14159)) 
print('{:06.2f}'.format(3.14159)) 
print('{:<6.2f}'.format(3.14159)) 
print('{:^6.2f}'.format(3.14159)) 
print('{:^e}'.format(3.14159)) 
print('{:.3e}'.format(3.14159))
print('{:.1e}'.format(3.14159))
```

# Formatting Strings

To format strings, the formatting code is {:s}

```python
print('{:^10s}'.format('Hi'))
print('{:10s}'.format('Hi'))
print('{:>10s}'.format('Hi'))
```
